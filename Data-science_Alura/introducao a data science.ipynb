```python
import pandas as pd
tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb.head()
tmdb.original_language.unique()
```

```python
tmdb.original_language.value_counts()
```

```python
cont_lingua = tmdb["original_language"].value_counts().to_frame().reset_index()
cont_lingua.columns = ["original_language", "total"]
cont_lingua.head()
```

```python
sns.barplot(x="original_language", y="total", data=cont_lingua)
```

```python
sns.catplot(x="original_language", kind="count", data = tmdb)
```

```python
print(sns.__version__)
```

```python
import seaborn as sns
total_lingua = tmdb.original_language.value_counts()
total_ingles = total_lingua.loc["en"]
total_geral = total_lingua.sum()
total_outros = total_geral - total_ingles
dados = {
    'Lingua': ['Ingles', 'Outros'],
    'Total': [total_ingles, total_outros]
}
dados = pd.DataFrame(dados)
print(dados)
sns.barplot(x = 'Lingua', y = 'Total', data = dados)
```

```python
filmes_n_en = tmdb.query("original_language != 'en'")
total_lingua_outros = filmes_n_en.original_language.value_counts()
dados = pd.DataFrame(total_lingua_outros).reset_index()
sns.barplot(x = 'index', y = 'original_language', data = dados)
```

```python
filmes_n_en = tmdb.query("original_language != 'en'")
sns.catplot(x="original_language", kind="count", data = filmes_n_en, aspect = 2, order = total_lingua_outros.index, palette = "GnBu_d",)
```
